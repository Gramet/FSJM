```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
# Function to plot the sine wave
def plot_schema(a_ab_ratio):
    R_m = 3
    R_h = 2
    omega_m = 1
    omega_h = 1 / 12

    t = np.linspace(0, 12 * 2 * np.pi, 500)

    x_m = R_m * np.sin(omega_m * t)
    y_m = R_m * np.cos(omega_m * t)

    x_h = R_h * np.sin(omega_h * t)
    y_h = R_h * np.cos(omega_h * t)

    x_p = x_h + a_ab_ratio * (x_m - x_h)
    y_p = y_h + a_ab_ratio * (y_m - y_h)

    plt.figure(figsize=(8, 4))
    fig, ax = plt.subplots()
    ax.plot(x_m, y_m, label="Minute circle")
    ax.plot(x_h, y_h, label="Hour circle")
    ax.plot(
        x_p,
        y_p,
        label=f"Black Point with a_b_ratio ratio {a_ab_ratio/(1-a_ab_ratio):.3f}",
    )
    ax.set_title("Problem 18")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.axhline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.legend(bbox_to_anchor=(1, 0.5))
    ax.set_box_aspect(1)
    ax.grid(True)
    fig.tight_layout()
    plt.show()


# Widget to control frequency
frequency_slider = widgets.FloatSlider(
    value=0.1, min=0.0, max=0.1, step=0.001, description="Frequency:"
)

# Interactive plot
interact(plot_schema, a_ab_ratio=frequency_slider)
```

```python
# Function to plot the sine wave
def plot_schema(a_b_ratio):
    R_m = 3
    R_h = 2
    omega_m = 1
    omega_h = 1 / 12

    t = np.linspace(0, 12 * 2 * np.pi, 500)

    x_m = R_m * np.sin(omega_m * t)
    y_m = R_m * np.cos(omega_m * t)

    x_h = R_h * np.sin(omega_h * t)
    y_h = R_h * np.cos(omega_h * t)

    a_ab_ratio = a_b_ratio / (a_b_ratio + 1)
    x_p = x_h + a_ab_ratio * (x_m - x_h)
    y_p = y_h + a_ab_ratio * (y_m - y_h)

    plt.figure(figsize=(8, 4))
    fig, ax = plt.subplots()
    ax.plot(x_m, y_m, label="Minute circle")
    ax.plot(x_h, y_h, label="Hour circle")
    ax.plot(
        x_p,
        y_p,
        label=f"Black Point with a_b_ratio ratio {a_b_ratio:.3f}",
    )
    ax.set_title("Problem 18")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.axhline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.legend(bbox_to_anchor=(1, 0.5))
    ax.set_box_aspect(1)
    ax.grid(True)
    fig.tight_layout()
    plt.show()


# Interactive plot
plot_schema(a_b_ratio=1 / 18)  # Actual solution
plot_schema(a_b_ratio=2 / 3)  # Degenerative solution
```
